```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
fig, ax=plt.subplots(2, 2)
ax[0, 1].plot(x, np.sin(x))
ax[1, 1].plot(x, np.cos(x))
plt.show()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)

model = lambda x: x * np.sin(x)
xd = np.array([1, 3, 5, 6, 8])
yd = model(xd)

gp = GaussianProcessRegressor()
gp.fit(xd[:, np.newaxis], yd)

xfit = np.linspace(0, 10, 1000)
yfit, dyfit_ori = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * dyfit_ori

fig = plt.figure()
ax = plt.axes()
ax.plot(xd, yd, 'o')
ax.plot(xfit, yfit)
# ax.plot(xfit, yfit-dyfit)
# ax.plot(xfit, yfit+dyfit)
plt.fill_between(xfit, yfit-dyfit, yfit+dyfit, color='gray', alpha=0.2)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

zl = np.linspace(0, 15, 1000)
xl = np.sin(zl)
yl = np.cos(zl)
ax.plot3D(xl, yl, zl, 'g')
zd = 15 * np.random.random(100)
xd = np.sin(zd) + 0.1 * np.random.randn(100)
yd = np.cos(zd) + 0.1 * np.random.randn(100)
ax.scatter3D(xd, yd, zd, c=zd, cmap='Greens');
```
